# Analisando negociações do Robo Max Min

```python
import pandas as pd
from pathlib import Path
```

```python
path_to_file = Path().cwd().parent / 'data' / 'trades_maxmin_win.xlsx'
```

```python
columns = ['Horario','Oferta','Ativo','Tipo','Direção', 'Volume','Preço','Ordem','Comissão','Swap','Lucro','Saldo']
df = pd.read_excel(path_to_file, sheet_name='Sheet2', header=None)
df.columns = columns
df.set_index('Horario', inplace=True)
df.head()
```

```python
df['Lucro'].cumsum().plot(figsize=(12,6), title='Curva de Capital', ylabel='Lucro');
```
